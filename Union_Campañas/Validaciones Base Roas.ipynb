```python
#Al traernos la información de google sheets no se respeta el formato, todo se mantiene en object
base_roas.costo_planeado = base_roas.costo_planeado.apply(lambda x : float(x))
base_roas.kpi_planeado = base_roas.kpi_planeado.apply(lambda x: str(x).replace(',','')).astype('int')
base_roas.serving = base_roas.serving.apply(lambda x: str(x).replace(',','')).astype('int')
base_roas.inversión_plataforma = base_roas.inversión_plataforma.apply(lambda x: str(x).replace(',','')).astype('int')
base_roas.inversión_total = base_roas.inversión_total.apply(lambda x: str(x).replace(',','')).astype('int')
```

```python
#La información del mp llega hasta la columna plan_x
mp = base_roas.loc[:,:'plan_x']
```

```python
enero_prueba = mp[(mp.cliente == 'GICSA') & (mp.marca == 'Forum Cuernavaca') & (mp.mes_plan == 'Enero') ]

enero_prueba
```

```python
enero_prueba.groupby(['cliente','marca','mes_plan','versión','plataforma_x'], as_index = False).agg({'costo_planeado':'mean',
                                                                                              'kpi_planeado':'mean',
                                                                                              'serving':'mean',
                                                                                              'inversión_plataforma':'mean',
                                                                                              'inversión_total':'mean'})
```
